```python
# Part 2
# Bond Length Calculation function
import math
def compute_bond_length(coord1, coord2):
    """
    Computes the bond lengths for a molecule given a set of coordinates.

    Parameters:
    coord1 (list): coordinates of a molecule.
    coord2 (list): coordinates of a molecule.

    Returns:
    float: The bond length.
    """
    # Check coord1 and coord2 are the same size
    if len(coord1) != len(coord2):
        print("ERROR in compute_bond_length coordinates")
    squaredSum = 0
    for i in range(len(coord1)):
        squaredSum += pow((coord2[i] - coord1[i]),2.0)
    sqrtSum = math.sqrt(squaredSum)
    # Check bond length is not too long
    if sqrtSum > 2:
        print("ERROR in compute_bond_length: bond length is too long")
    return sqrtSum
# Test compute_bond_length
# Turn on / off test
TEST = True
if TEST:
    print("COMPUTE_BOND_LENGTH FUNCTION TESTING")
    print("Benzene test:")
    print(compute_bond_length(Benzene["C1"], Benzene["C2"]))
    print("Hydrogen Gas Test:")
    print(compute_bond_length(H2["H1"], H2["H2"]))
```

```python
# Part 3
import numpy as np
# Helper functions
# Vector Maker function
def vectorMaker(coord1, coord2):
    # Check coord1 and coord2 are the same size
    if len(coord1) != len(coord2):
        print("ERROR in vectorMaker coordinates")
    vector = np.zeros(len(coord1))
    for i in range(len(vector)):
        vector[i] = coord2[i] - coord1[i]
    return vector
# Test vectorMaker
if TEST:
    print("vectorMaker FUNCTION TESTING")
    print("Benzene test:")
    print(vectorMaker(Benzene["C1"], Benzene["C2"]))
    print("Hydrogen Gas Test:")
    print(vectorMaker(H2["H1"], H2["H2"]))
def dotProd(vec1, vec2):
    # Check vec1 and vec2 are the same size
    if len(vec1) != len(vec2):
        print("ERROR in dotProd vectors")
    vecSum = 0
    for i in range(len(vec1)):
        vecSum += vec1[i] * vec2[i]
    return vecSum
# Test dotProd
if TEST:
    print("dotProd FUNCTION TESTING")
    testvec1 = vectorMaker(Benzene["C1"], Benzene["C2"])
    testvec2 = vectorMaker(Benzene["C2"], Benzene["C3"])
    print(dotProd(testvec1, testvec2))
def vecLength(vec1):
    vecSum = 0
    for i in range(len(vec1)):
        vecSum += pow(vec1[i],2.0)
    return math.sqrt(vecSum)
# Test vecLength
if TEST:
    print("vecLength FUNCITON TESTING")
    testvec1 = vectorMaker(Benzene["C1"], Benzene["C2"])
    testvec2 = vectorMaker(Benzene["C2"], Benzene["C3"])
    print(vecLength(testvec1))
    print(vecLength(testvec2))
# Bond Angle Calculation function
def compute_bond_angle(coord1, coord2, coord3):
    """
    Computes the bond angles for a molecule given a set of coordinates.

    Parameters:
    coord1 (list): coordinates of a molecule.
    coord2 (list): coordinates of a molecule.
    coord3 (list): coordinates of a molecule.

    Returns:
    float: The bond angle.
    """
    AB = vectorMaker(coord1, coord2)
    BC = vectorMaker(coord2, coord3)
    return math.degrees(math.acos((dotProd(AB,BC))/(vecLength(AB)*vecLength(BC))))
# Test compute_bond_angle
if TEST:
    print("compute_bond_angle FUNCTION TESTING")
    print(compute_bond_angle(Benzene["C1"], Benzene["C2"], Benzene["C3"]))
```

```python
# Part 4
# Part 4.1
# Calculate all bond lengths function
def calculate_all_bond_lengths(molecule):
    """
    Computes the bond lengths for a given molecule.

    Parameters:
    molecule (dictionary): molecule in question.

    Returns:
    list: The bond lengths.
    """
    bond_lengths = []
    for atom1 in molecule:
        for atom2 in molecule:
            if atom1 != atom2:
                # Compute bond length only if it's a unique pair
                bond_lengths.append(compute_bond_length(molecule[atom1], molecule[atom2]))
    return bond_lengths
# Test calculate_all_bond_lengths
if TEST:
    print("calculate_all_bond_lengths FUNCTION TESTING")
    print(calculate_all_bond_lengths(Benzene))
```

```python
# Part 4.2
# Calculate all bond angles
def calculate_all_bond_angles(molecule):
    """
    Computes the bond angles for a given molecule.

    Parameters:
    molecule (dictionary): molecule in question.

    Returns:
    list: The bond angles.
    """
    bond_angles = []
    for atom1 in molecule:
        for atom2 in molecule:
            if atom1 != atom2:
                for atom3 in molecule:
                    if atom2 != atom3 and atom3 != atom1:
                        bond_angles.append(compute_bond_angle(molecule[atom1], molecule[atom2], molecule[atom3]))
    return bond_angles
# Test calculate_all_bond_angles
if TEST:
    print("calculate_all_bond_angles FUNCTION TESTING")
    print(calculate_all_bond_angles(Benzene))
```
