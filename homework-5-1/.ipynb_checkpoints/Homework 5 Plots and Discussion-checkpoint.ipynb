```python
# Imports
from hydrogen_orbital_functions import *
import matplotlib.pyplot as plt
```

```python
# Plot for 1.1.2
integral = random_sampling_integral()

plt.xscale('log')
plt.plot([int(math.pow(10, i)) for i in range(2,8)], integral)
plt.xlabel("Number of Sampled Points")
plt.ylabel("Integral Value")
plt.title("Integral for Two H 2p Orbitals")
plt.show()
```

The Integral Plot converges fairly quickly to ~1.81, even being at 1.84 at 10 to the fourth power. It's also interesting that 10 to the second power is so far off in the beginning, but it's likely this is just due to the power putting in exponentially many points.
