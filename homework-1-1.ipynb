```python
# PART A

# Import Pandas
import pandas as pd

# Import the volume and pressure data in Pandas
df = pd.read_csv("volume_pressure_data.csv")

# Print the first few rows of the dataset
print("First few rows of Volume and Pressure data:")
print(df[:3])

# Print Mean
print("\nMean: ")
print(df.mean())

# Print Median
print("\nMedian: ")
print(df.median())

# Print Standard Deviation
print("\nStandard Deviation: ")
print(df.std())
```

```python
# Part B

# Import Matplotlib
import matplotlib.pyplot as plt
# Plot the data set
df.plot(x="Volume",y="Pressure")
# Title the plot
plt.title("Volume v.s. Pressure")
# Label y-axis
plt.ylabel("Pressure ($Hg$)")
# Label x-axis
plt.xlabel("Volume ($in^3$)")
# Add gridlines
plt.grid(True)
# Display the plot
plt.show()
```
